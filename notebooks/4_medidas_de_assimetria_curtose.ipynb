# Medidas de assimetria e curtose

## Assimetria

Assimetria, como o próprio nome insinua, significa desvio ou afastamentoo da simetria. Em outras palavras, assimetria é o grau de deformação de uma
curva de frequências. Quanto ao grau de deformação ou assimetria, podemos ter três tipos de curvas de frequências, a saber:

* Curva *simétrica*: $\bar{x} \approx  \tilde{x} \approx Mo$
* Curva *assimétrica à direita*: $\bar{x} > \tilde{x} > Mo$
* Curva *assimétrica à esquerda*: $\bar{x} < \tilde{x} < Mo$

A medida de *assimetria de Pearson* é dada por:

$$g = \frac {\bar{x} - Mo} {s}$$

*Propriedade: $g$ é adimensional.*

Como o cálculo da moda nem sempre é possível, uma possível proposta seria:

$$g = \frac {\bar{x} - \tilde{x}} {s}$$


## Curtose

A curtose, ou excesso, indica até que ponto a curva de frequências de uma distribuição se apresenta mais afilada ou mais achatada do que uma curva
padrão, denominada curva normal (sino). Em outras palavras, a curtose é o grau de achatamento de uma curva de frequência de uma distribuição. Quanto ao grau de achatamento, podemos ter três tipos de curvas de frequências, a saber:

- Curva *mesocúrtica* [$C \approx 0$]: achatamento da distribuição normal
- Curva *leptocúrtica* [$C > 0$]: distribuição mais concentrada
- Curva *platicúrtica* [$C < 0$]: distribuição mais achatada

A medida de curtose (baseada em momentos centrais amostrais) é dada por:

$$ g_{2} = \frac {m_{4}} {m_{2}^{2}} = \frac {\frac{1}{n}\sum \limits _{i=1} ^{n} (x_{i} - \bar{x})^{4}} {\Biggr[ \frac{1}{n}\sum \limits _{i=1} ^{n} (x_{i} - \bar{x})^{2}\Biggr]^{2}}$$

$$ C = g_{2} - 3 $$

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

from scipy.stats import kurtosis
```

```python
x = np.linspace(-5, 5, 100)
ax = plt.subplot()

for dist in ['laplace', 'norm', 'uniform']:
    if dist == 'uniform':
        d = getattr(stats, dist)(loc=-2, scale=4)
    else:
        d = getattr(stats, dist)
    data = d.rvs(size=1000)
    kur = kurtosis(data, fisher=True)
    y = d.pdf(x)
    ax.plot(x, y, label="{}, kur={}".format(dist, round(kur, 3)))
    ax.legend()
```
